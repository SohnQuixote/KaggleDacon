```python
lunch_train = train.drop('석식계' , axis = 1)
lunch_train.head()
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
folds=[]
for train_idx, valid_idx in skf.split(lunch_train, lunch_train['중식계']):
    folds.append((train_idx, valid_idx))
```

```python
from lightgbm import LGBMRegressor
random.seed(42)
lgb_models={}
learning_rate_array = []
for fold in range(5) :
    print(f'===================================={fold+1}============================================')
    train_idx, valid_idx = folds[fold]
    X_train, X_valid, y_train, y_valid = lunch_train.drop(['중식계'],axis=1).iloc[train_idx].values, lunch_train.drop(['중식계'],axis=1).iloc[valid_idx].values,\
                                         lunch_train['중식계'][train_idx].values, lunch_train['중식계'][valid_idx].values 
    lgb = LGBMRegressor(objective='quantile', alpha=fold,
                         n_estimators=10000, bagging_fraction=0.7, learning_rate=0.027, subsample=0.7)
    lgb.fit(X_train, y_train, 
            eval_set=[(X_train, y_train), (X_valid, y_valid)], 
            early_stopping_rounds=300,
           verbose=500)
    #temp_array = lgb.predict(X_valid)
    #learning_rate_array.append(cmp(temp_array,y_valid))
    lgb_models[fold]=lgb
    print(f'================================================================================\n\n')
```
